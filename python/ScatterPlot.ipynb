# 散点图

本篇内容主要介绍如何利用python画散点图。 个人觉得matplotlib自带的画图不是很友好，所以偏向于使用seaborn包画图

```python
import seaborn as sns
from matplotlib import pyplot as plt
%matplotlib inline 
import pandas as pd
```

```python
iris = sns.load_dataset("iris")
```

```python
print(iris.head())
```

```python
f, ax = plt.subplots(figsize=(7,5))
sns.scatterplot(data=iris, x='sepal_length', y='sepal_width',hue='species', style='species', ax=ax)
ax.set_title('Seaborn Scatter Plot')
ax.set_xlabel('Sepal Length')
ax.set_ylabel('Sepal Width')
```

参数解析:
* data: 要画图的数据集
* x: 作为x轴的数据的列名
* y: 作为y轴的数据的列名
* hue: 颜色标识
* style: 形状标识

### 删除图例中的标题

```python
f, ax = plt.subplots(figsize=(7,5))
sns.scatterplot(data=iris, x='sepal_length', y='sepal_width',hue='species', style='species', ax=ax)
ax.set_title('Seaborn Scatter Plot')
ax.set_xlabel('Sepal Length')
ax.set_ylabel('Sepal Width')
handles, labels = ax.get_legend_handles_labels()
ax.legend(handles=handles[1:], labels=labels[1:], title='Whatever you like')
```
